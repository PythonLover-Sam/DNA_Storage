```python
class SyntheticSequences:
    def __init__(self, fpo_index, fpi_index, vfp_index, payload_index, vrp, rpi_index, rpo_index, hd, seq_name):
        #              外侧引物        内侧      特征         载荷      特征下  内引        外引 非特异性距离  序列名称
        #  注意：初始化序列时不需要指定非特异性引物结合位点，只需要输入原始引物和非特异性距离即可
        self.dna = ''
        self.seq_name = seq_name
        self.fpo = fps[fpo_index]
        self.fpi = fps[fpi_index]
        self.vfp = all_f_vectors[vfp_index]
        self.payload = all_f_seqs[payload_index]
        self.vrp = vrp
        self.rpi = rps[rpi_index]
        self.rpo = rps[rpo_index]
        self.hd = hd
        self.fpi_idx = fpi_index
        self.rpi_idx = rpi_index
        self.fpo_idx = fpo_index
        self.rpo_idx = rpo_index
        self.dna_structure = ''
        
        if hd == 0:
            self.dna_structure = strand.assemble_primer_and_payload([fps[fpo_index], fp0s[fpi_index], self.vfp], self.payload, [self.vrp, rp0s[rpi_index], rps[rpo_index]])
            self.dna = self.dna_structure[0]
            self.seq_name += "_0HD"
        elif hd == 2:
            self.dna_structure = strand.assemble_primer_and_payload([fps[fpo_index], fp2s[fpi_index], self.vfp], self.payload, [self.vrp, rp2s[rpi_index], rps[rpo_index]])
            self.dna = self.dna_structure[0]
            self.seq_name += "_2HD"
        elif hd == 4:
            self.dna_structure = strand.assemble_primer_and_payload([fps[fpo_index], fp4s[fpi_index], self.vfp], self.payload, [self.vrp, rp4s[rpi_index], rps[rpo_index]])
            self.dna = self.dna_structure[0]
            self.seq_name += "_4HD"
        elif hd == 6:
            self.dna_structure = strand.assemble_primer_and_payload([fps[fpo_index], fp6s[fpi_index], self.vfp], self.payload, [self.vrp, rp6s[rpi_index], rps[rpo_index]])
            self.dna = self.dna_structure[0]
            self.seq_name += "_6HD"
        else:
            raise ValueError("hd只能为0，2，4，6")

    def check_pcr(self, seq):
        strand.check_max_2seqs_complement_hybird(seq, self.dna)
        strand.check_max_2seqs_complement_hybird(seq, strand.get_complement_strand(self.dna))
        
        return self
    def check_binding_site(self, seq):
        # 检查给定序列与有效载荷部分的杂交情况
        strand.check_max_2seqs_complement_hybird(seq, self.payload)
        strand.check_max_2seqs_complement_hybird(seq, strand.get_complement_strand(self.payload))
        
        return self
```
